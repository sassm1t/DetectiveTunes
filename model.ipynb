```python
import pandas as pd
df = pd.read_csv("data.csv")
df
df1=df[df['sourcename']!=df['targetname']]
df1
```

```python
duplicate = []    #this piece of code to add a new column to the dataframe 
#which will contain the 0 or 1 value to check whether the piece is plagiarised based on
#cosine similarity measurement
for i in df.score:
   if i > 0.5:  #here the plagiarism threshold value is adjusted for 0.8,0.7,0.6,0.5
      duplicate.append(1)
   else:
      duplicate.append(0)

df['duplicate'] = duplicate
```

```python
df
```

```python
df = df.sample(frac=1).reset_index(drop=True)
df
```

```python
import re
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
corpus = []   #the chords are converted to bag of words to be
#vectorized

for i in range(len(df)):
    source = re.sub("|\,|\'|,","",df['sourceChord'][i])
    target = re.sub("|\,|\'|,","",df['targetChord'][i])
    source = source.lower()
    target = target.lower()
    source = source.split()
    target = target.split()
    source = ' '.join(source)
    target = ' '.join(target)
    sourceandtarget = source + target
    corpus.append(sourceandtarget)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
cv = CountVectorizer()
X = cv.fit_transform(corpus).toarray()  #vectorization happens here
```

```python
y = df['duplicate']    #train test spliting happens

from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state = 0)
```

```python
#Random forest classifier
from sklearn.ensemble import RandomForestClassifier
classifier3 = RandomForestClassifier(n_estimators = 10, criterion = 'entropy', random_state = 0)
classifier3.fit(X_train, y_train)
     
```

```python
y_pred = classifier3.predict(X_test)
```

```python
import matplotlib.pyplot as plt
# Visualise classical Confusion M0atrix
from sklearn import metrics
from sklearn.metrics import confusion_matrix, classification_report
CM = confusion_matrix(y_test, y_pred)
cm_display = metrics.ConfusionMatrixDisplay(confusion_matrix = CM, display_labels = [False, True])
cm_display.plot()
plt.show()

```
